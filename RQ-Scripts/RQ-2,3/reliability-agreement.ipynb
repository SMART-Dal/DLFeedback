```python
import numpy as np
import pandas as pd
import krippendorff
# run "pip install krippendorff" first.
```

```python
reliability_file = "matrix-output-before.csv"
file_smell_mapping_file = "output_before.csv"
```

```python
reliability_df = pd.read_csv(reliability_file)
```

```python
reliability_df
```

```python
reliability_matrix = reliability_df.T
```

```python
reliability_matrix.head()
```

```python
reliability_matrix = reliability_matrix.iloc[1:, :]
```

```python
reliability_matrix
```

```python
reliability_matrix = reliability_matrix.values.tolist()
```

```python
reliability_matrix = [[np.nan if val == "-" else 1 if val == "T" else 0 for val in sublist] for sublist in reliability_matrix]
```

```python
len(reliability_matrix)
```

```python
overall_alpha = krippendorff.alpha(reliability_data=reliability_matrix, level_of_measurement="nominal")
print(f"Overall Krippendroff's alpha: {overall_alpha}")
```

```python
# Calculating Krippendorff's alpha at smell level
file_smell_df = pd.read_csv(file_smell_mapping_file)
```

```python
file_smell_df["Smell"].unique()
```

```python
grouped_by_smell = file_smell_df.groupby("Smell")
```

```python
mapping = {}
```

```python
for val, group in grouped_by_smell:
    mapping[val] = group["File"].unique()
```

```python
reliability_df
```

```python
file_smell_df
```

```python
reliability_df = reliability_df.rename(columns={'fileName': 'File'})
```

```python
reliability_df
```

```python
merged_df = pd.merge(reliability_df, file_smell_df, on='File')
```

```python
merged_df
```

```python
merged_df = merged_df.drop_duplicates(subset=["File"])
```

```python
merged_df
```

```python
cm_df = merged_df[merged_df["Smell"] == "ComplexMethod"].drop(columns=["UserId", "Smell", "ModelPrediction", "UserAnnotation"])
lpm_df = merged_df[merged_df["Smell"] == "LongMethod"].drop(columns=["UserId", "Smell", "ModelPrediction", "UserAnnotation"])
mfa_df = merged_df[merged_df["Smell"] == "MultiFaceted"].drop(columns=["UserId", "Smell", "ModelPrediction", "UserAnnotation"])
```

```python
cm_df = cm_df.T.iloc[1:, :]
```

```python
lpm_df = lpm_df.T.iloc[1:, :]
```

```python
mfa_df = mfa_df.T.iloc[1:, :]
```

```python
cm_df = cm_df.values.tolist()
lpm_df = lpm_df.values.tolist()
mfa_df = mfa_df.values.tolist()
```

```python
cm_reliability_matrix = [[np.nan if val == "-" else 1 if val == "T" else 0 for val in sublist] for sublist in cm_df]
```

```python
lpm_reliability_matrix = [[np.nan if val == "-" else 1 if val == "T" else 0 for val in sublist] for sublist in lpm_df]
```

```python
mfa_reliability_matrix = [[np.nan if val == "-" else 1 if val == "T" else 0 for val in sublist] for sublist in mfa_df]
```

```python
cm_alpha = krippendorff.alpha(reliability_data=cm_reliability_matrix, level_of_measurement="nominal")
```

```python
lpm_alpha = krippendorff.alpha(reliability_data=lpm_reliability_matrix, level_of_measurement="nominal")
```

```python
#!!!
# mfa_alpha = krippendorff.alpha(reliability_data=mfa_reliability_matrix, level_of_measurement="nominal")
```

```python
print(f"ComplexMethod Krippendroff's alpha: {cm_alpha}")
```

```python
print(f"LPM Krippendroff's alpha: {lpm_alpha}")
```
