```python
import os

from sklearn.metrics import cohen_kappa_score
import numpy as np
from scipy.stats import gmean
```

```python
folder_path_before = "C:\\Users\\author\\Documents\\thesis\\plugin\\code\\TagcoderDataTransformer\\FilesForKappaPython\\Before\\"
folder_path_after = "C:\\Users\\author\\Documents\\thesis\\plugin\\code\\TagcoderDataTransformer\\FilesForKappaPython\\After\\"
```

```python
file_list_before = os.listdir(folder_path_before)
file_list_after = os.listdir(folder_path_before)
```

```python
for file_name in file_list_before:
        file_contents = []
        file_path = os.path.join(folder_path_before, file_name)

        kappa_scores = []

        if os.path.isfile(file_path):
            # Open the file and read its contents
            with open(file_path, 'r') as file:
                lines = file.readlines()
                line1 = lines[0].strip().split(',')
                line2 = lines[1].strip().split(',')

                line1 = [1 if val == 'T' else 0 for val in line1]
                line2 = [1 if val == 'T' else 0 for val in line2]

                if line1 and line2:
                # print(line1)
                # print(line2)
                    print('\n')
                    kappa_score = cohen_kappa_score(line1, line2)
                    print('kappa score : ',kappa_score)
                    kappa_scores.append(kappa_score)

mean_kappa_score = np.mean(kappa_scores)
geometric_mean = gmean(kappa_scores)
print("Mean Cohen's kappa score:", geometric_mean)
        
```
