```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
import seaborn as sns
plt.style.use(['science', 'ieee'])
```

```python
# Load the data from CSV file
data = pd.read_csv('../results/participants.csv',encoding = 'unicode_escape',header=0)
```

```python
data
```

```python
data = data.rename(columns={'Before': 'Round 1', 'After': 'Round 2'})
```

```python
data['Group'] = data['Group'].replace({'A': 'Experiment', 'P': 'Control'})
```

```python
data
```

```python
df = data
```

```python
melted_df = pd.melt(df, id_vars=['Participant', 'Total-Size', 'Group'], value_vars=['Round 1', 'Round 2'], var_name='Round', value_name='Score')
```

```python
melted_df
```

```python
plt.figure(figsize=(6, 4))
ax = sns.violinplot(data=melted_df, x="Round", y="Score", hue="Group",
               inner="quart", linewidth=.8,  palette="tab10")
plt.rc('axes', labelsize=25)
plt.rc('xtick', labelsize=25)
plt.rc('ytick', labelsize=25)
plt.rc('legend', fontsize=15)
plt.tight_layout()
ax.legend(loc='lower right')
plt.savefig("f1_scores_violin.pdf")
```
